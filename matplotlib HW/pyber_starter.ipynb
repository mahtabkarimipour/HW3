```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
city_data = pd.read_csv('data/city_data.csv')
ride_data = pd.read_csv('data/ride_data.csv')
# Combine the data into a single dataset
data = pd.merge(city_data,ride_data ,
                                 how='outer', on='city')
data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_group = data.groupby(['city'])
mean_city = city_group['fare'].mean()
driver_city = city_group['driver_count'].sum()
ride_city = city_group['ride_id'].count()

# Build the scatter plots for each city types


# Incorporate the other graph properties

# Create a legend
# Incorporate a text label regarding circle size

num_of_rural = city_data.loc[city_data.type == 'Rural', 'type'].count()
num_of_urban = city_data.loc[city_data.type == 'Urban', 'type'].count()
num_of_suburban = city_data.loc[city_data.type == 'Suburban', 'type'].count()
suburban_color = ["Lightskyblue"] * num_of_suburban
urban_color = ["Lightcoral"] * num_of_urban
rural_color = ["Gold"] * num_of_rural
color = rural_color + suburban_color + urban_color
size = [(i)/5 for i in driver_city.tolist()]
plt.scatter(ride_city, mean_city, marker="o", c=color, alpha=0.75, s=size)
plt.title("Piber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fair ($)")
```

## Total Fares by City Type

```python
# Calculate Type Percents
x=(data[data['type'] == 'Rural']['fare'].sum() / data['fare'].sum() )*100 
y=(data[data['type'] == 'Urban']['fare'].sum() / data['fare'].sum() )*100 
z=(data[data['type'] == 'Suburban']['fare'].sum() / data['fare'].sum() )*100 
r=[x,y,z]
# Build Pie Chart
labels = ["Rural", "Urban", "Suburban"]
colors = ["Gold", "Lightcoral", "Lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(r, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% of total fare by city type")
```

## Total Rides by City Type

```python
x=(data[data['type'] == 'Rural']['ride_id'].count() / data['driver_count'].count() )*100 
y=(data[data['type'] == 'Urban']['ride_id'].count() / data['driver_count'].count() )*100 
z=(data[data['type'] == 'Suburban']['ride_id'].count() / data['driver_count'].count() )*100 
r=[x,y,z]

# Build Pie Chart
labels = ["Rural", "Urban", "Suburban"]
colors = ["Gold", "Lightcoral", "Lightskyblue"]
explode = (0.1, 0, 0)

plt.pie(r, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% of total rides by city type")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
x=(data[data['type'] == 'Rural']['driver_count'].sum() / data['driver_count'].sum() )*100 
y=(data[data['type'] == 'Urban']['driver_count'].sum() / data['driver_count'].sum() )*100 
z=(data[data['type'] == 'Suburban']['driver_count'].sum() / data['driver_count'].sum() )*100 
r=[x,y,z]

# Build Pie Chart
labels = ["Rural", "Urban", "Suburban"]
colors = ["Gold", "Lightcoral", "Lightskyblue"]
explode = (0.1, 0, 0)

plt.pie(r, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% of total Drivers by city type")
```
